```python
#importing the packages

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image= cv2.imread('MicrosoftTeams-image.png')
image= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)


fig= plt.figure(figsize= (20,10))
plt.imshow(image)
```

```python
image= cv2.imread('MicrosoftTeams-image.png')
image= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
fig= plt.figure(figsize= (20,10))
image= cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.imshow(image,cmap='gray')
```

```python
# load cascade classifier
face_cascade= cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')

faces = face_cascade.detectMultiScale(image,1.3,5) #black/white img, ratio value,window size 
print(faces)
print('no of faces', len(faces))
```

```python
image= cv2.imread('MicrosoftTeams-image.png')
image= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
for (x,y,w,h) in faces:
    image= cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),3) #img,x,y, final value(x+w,y+h), bounding box color(rgb), width
fig= plt.figure(figsize= (20,10))
plt.imshow(image)
```
